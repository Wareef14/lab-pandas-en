### 3 - In this part of the lab, we would like to test complex conditions on the entire data set at once. Let's start by finding the number of rows where the CGPA is greater than 9 and the student has performed an investigation.

```python
# Filter rows where CGPA > 9 and Research == 1
filtered_data = admissions[(admissions['CGPA'] > 9) & (admissions['Research'] == 1)]

# Count the number of rows that meet the conditions
num_rows = len(filtered_data)

print(f"Number of rows where CGPA > 9 and Research == 1: {num_rows}")
```

```python
specific_row = filtered_data.iloc[100]
print(specific_row)
```

### 4 - Now return all the rows where the CGPA is greater than 9 and the SOP score is less than 3.5. Find the mean chance of admit for these applicants.

```python
# Filter rows where CGPA > 9 and SOP < 3.5
filtered_data = admissions[(admissions['CGPA'] > 9) & (admissions['SOP'] < 3.5)]

# Display the filtered rows
filtered_data
```

```python
# Apply the condition (CGPA > 9 and SOP < 3.5)
filtered_rows = admissions[(admissions['CGPA'] > 9) & (admissions['SOP'] < 3.5)]

# Calculate the mean 'Chance of Admit' for these applicants
mean_chance_of_admit = filtered_rows['Chance of Admit '].mean()

print(mean_chance_of_admit)
```
